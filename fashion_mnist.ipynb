```
import tensorflow as tf
import numpy as np
```

```
tf.__version__
```

```
(x_train , y_train) , (x_test , y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```
x_train.shape
```

```
28 * 28
```

```
np.max(x_train[:])
```

```
x_train = x_train / 255.
x_test  = x_test / 255.
```

```
x_train = x_train.reshape(-1,784)
x_test = x_test.reshape(-1,784)
```

```
x_train.shape
```

```
model = tf.keras.Sequential()
```

```
model.add(tf.keras.layers.Dense(units=392 , activation='relu' , input_shape=(784,)))
model.add(tf.keras.layers.Dense(units=196 , activation='relu' , input_shape=(392,)))
model.add(tf.keras.layers.Dense(units=10, activation='softmax' , input_shape=(196,)))
```

```
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy' , metrics=['sparse_categorical_accuracy'])
```

```
model.summary()
```

```
model.fit(x_train,y_train, epochs=7)
```

```
test_loss , test_accurency = model.evaluate(x_test,y_test)
print('Presicion: ',test_accurency)
print('Perdida: ',test_loss)
```
