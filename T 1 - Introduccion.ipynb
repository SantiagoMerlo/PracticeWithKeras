## Constantes

### Definicion de constantes
* Entender como crear variables constantes
* La principal ventaja comparada con TF1.X es la comodidad que se puede obtener con la integracion de numpy
* Los objetos de tensorflow requieren arrays

```
tensor = tf.constant([[20,4] ,[32,52]])
tensor
```

```
#se puede ver los atributos tambien de un tensor
print(tensor.shape)
print(tensor.dtype)
print(tensor.numpy())
```

```
numpy_tensor = np.array([[25,4],[21,54]])
tensor_from_np = tf.constant(numpy_tensor)
tensor_from_np
```

## Variables

### Definicion de variables
* Se declaran con la clase `Variable`
* Se utilizan los mismos metodos para ver sus atributos
* La diferencia entre constantes y variables, es que la primera se mantiene igual en todo el modelo mientras el segundo puede ir variando

```
variable = tf.Variable([[1.,2.,3.],[4.,5.,6.]])
variable
```

### Cambiar la configuracion

Se puede modificar los valores e incluso obtenerlos igual que un arreglo comun y corriente

```
variable[0,2].assign(100)
```

### Operacion con los tensores
* Al estar utilizando una matriz, se aplica suma escalar y multiplicacion escalar.
  * Basicamente es hacer un calculo a todos los elementos
  * Esto nos permite hacer calculos de algebra lineal
* La integracion con numpy es tal que nos permite aplicar todas los calculos que haciamos en numpy a nuestros tensores

```
tensor.numpy()
```

```
tensor + 2
```

```
tensor * 5
```

```
#Esto esta aplicando una multiplicacion por fila
tensor * [1,2]
```

```
print('obtener la raiz cuadrada: ',np.sqrt(tensor))
print('obtener el promedio: ',np.mean(tensor))
print('Obtener el cuadrado: ',np.square(tensor))
```

### Producto escalar entre tensor o producto tensorial
* Con esta hay que tener cuidado con esto, porque no es multiplicacion de tensor1 * tensor2 sino que es un producto escalar
* En otras palabras se aplica un producto matricial

```
tensor1 = tf.constant([[2,4],[2,4]])
tensor2 = tf.constant([[1,3],[1,3]])
tensor3 = np.dot(tensor1,tensor2)
print('Tensor 1: \n',tensor1.numpy())
print('Tensor 2: \n',tensor2.numpy())
print('Tensor 3: \n',tensor3)
```
